Menangani Data yang Tidak Konsisten

```python
print(df_mysql['nim'].head())
print(df_excel['ID number'].head())
```

```python
# Ubah ID number dijsdikan integer karena akan membuang angka decimal diakhir
df_excel['ID number'] = df_excel['ID number'].fillna(0).astype(int).astype(str)
df_excel.head()

# Tambahkan "0" didepan ID number : pola dari nim
df_excel['ID number'] = "0" + df_excel['ID number']
df_excel.head()
```

```python
# Standdarisasi tipe data. Agar aman samakan tipe data nya dahulu, dari excel ID number

df_excel = df_excel.copy()
df_mysql = df_mysql.copy()

df_excel['ID number'] = df_excel['ID number'].astype(str)
df_mysql['nim'] = df_mysql['nim'].astype(str)
```

```python
# Integrasi (penggabungan data)

df_integrasi = pd.merge(
    df_mysql,
    df_excel,
    left_on='nim',
    right_on='ID number',
    how='inner'
)

df_integrasi.head()
```

```python
# buat dataframe agar lebih rapi
df_integrasi = df_integrasi[['Last name', 'First name', 'ID number', 'Grade/100.00']]

df_integrasi.head()
```

```python
# renmae Grade/100.00 menjadi nilai

df_integrasi.rename(columns={'Grade/100.00': 'nilai'}, inplace=True)
df_integrasi.head()
```

```python
df_integrasi.info()
```

```python
# rename kolom 'First name' dan 'Last name' menjadi 'Nama lengkap'

df_integrasi['Nama lengkap'] = df_integrasi['First name'] + ' ' + df_integrasi['Last name']
df_integrasi.head()
```

```python
# Rename kolom 'ID number' menjadi 'nim'
df_integrasi.rename(columns={'ID number': 'nim'}, inplace=True)
df_integrasi.head()
```

```python
df_integrasi = df_integrasi[['Nama lengkap', 'nim', 'nilai']]
df_integrasi.head()
```
