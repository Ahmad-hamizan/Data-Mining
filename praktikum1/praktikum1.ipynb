### 6. Filtering Data

```python
# mengambil data yang hanya di propinsi DKI Jakarta
df_jakarta = df[df['propinsi'] == 'DKI Jakarta']
df_jakarta.head()
```

```python
# mengambil data yang hanya total tempat tidur lebih dari 100
df_RSbesar = df[df['median_total_tempat_tidur'] > 100]
df_RSbesar.head()
```

```python
# menagmbil data yang hanya di propinsi DKI Jakarta dan total tempat tidur lebih dari 100
df_filter = df[(df['propinsi'] == 'DKI Jakarta') & (df['median_total_tempat_tidur'] > 100)]
df_filter.head()
```

```python
# Mengambil data yang hanya di propinsi DKI Jakarta atau total tempat tidur kurang dari 10
df_filter_RSKecil = df[(df['propinsi'] == 'Papua') | (df['median_total_tempat_tidur'] < 10)]
df_filter_RSKecil.head()
```

```python
# mengambil data untuk menampilkan baris dengan kolom kelas yang bernilai A atau B
df_kelas = df[(df['kelas'] == 'A') | (df['kelas'] == 'B')]
df_kelas.head()
```

```python
# mengambil data kolom jenis yang mengandung kata "Jantung"
df[df['jenis'].str.contains('Jantung')]
```
